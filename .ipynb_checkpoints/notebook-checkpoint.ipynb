## Visualization & Explanatory Analysis

### Bagaimana pengaruh hari kerja terhadap jumlah penyewaan sepeda di berbagai musim?


```python
workingday_count_season = df_hour.groupby(["workingday","season"])["cnt"].sum().reset_index()
workingday_count_season["workingday"] = workingday_count_season["workingday"].replace({0 : 'No',1 : 'Yes'})
workingday_count_season["season"] = workingday_count_season["season"].replace({1 : 'Springer',2 : 'Summer',3 : 'Fall',4 : 'Winter'})
plt.figure(figsize=(10,6))
sns.barplot(data=workingday_count_season, x="season", y="cnt", hue="workingday", palette="Set2")
plt.show()
```

### Bagaimana pengaruh kondisi cuaca (weathersit) terhadap jumlah penyewaan sepeda di berbagai musim?


```python
weathersit_count_season = df_hour.groupby(["weathersit","season"])["cnt"].sum().reset_index()
weathersit_count_season["weathersit"] = weathersit_count_season["weathersit"].replace({1 : 'Clear',2 : 'Mist',3 : 'Light Snow',4 : 'Heavy Snow'})
weathersit_count_season["season"] = weathersit_count_season["season"].replace({1 : 'Springer',2 : 'Summer',3 : 'Fall',4 : 'Winter'})
plt.figure(figsize=(10,6))
sns.barplot(data=weathersit_count_season, x="season", y="cnt", hue="weathersit", palette="Set2")
plt.show()
```

### bagaimana perbandingan tren jumlah penyewaan sepeda dari tahun ke tahun?

```python
df_2011 = df_day[df_day['yr'] == 2011]
df_2012 = df_day[df_day['yr'] == 2012]
plt.figure(figsize=(20,6))
plt.plot(df_2011['dteday'], df_2011['cnt'], label='2011', color='blue')
plt.plot(df_2012['dteday'], df_2012['cnt'], label='2012', color='green')
plt.title('Jumlah Penyewaan Sepeda dari Tahun ke Tahun')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Penyewaan')
plt.legend()
plt.grid(True)
plt.show()
```

### bagaimana perbandingan tren jumlah penyewaan sepedaÂ berdasarkan jam?

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_hour, x='hr', y='cnt', marker='o',hue='yr')
plt.title('Jumlah Penyewaan Sepeda dari Jam ')
plt.xlabel('Jam')
plt.ylabel('Jumlah Penyewaan')
plt.xticks(ticks=range(24))
plt.grid()
plt.show()
```

### bagaimana perbandingan tren jumlah penyewaan sepeda berdasarkan klasifikasi penyewa?

```python
plt.figure(figsize=(20,6))
plt.plot(df_day['dteday'], df_day['casual'], label='Casual', color='blue')
plt.plot(df_day['dteday'], df_day['registered'], label='Registered', color='green')
plt.title('Tren Penyewaan Sepeda Berdasarkan Klasifikasi Penyewa')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Penyewaan')
plt.legend()
plt.grid(True)
plt.show()
```

## Conclusion
Setiap tahun, jumlah penggunaan sepeda dalam sistem bike sharing mengalami peningkatan yang cukup signifikan. Berdasarkan data, waktu penyewaan tertinggi terjadi pada pukul 8 pagi dan 5 sore, dengan mayoritas pengguna adalah pelanggan terdaftar. Penyewaan juga lebih banyak terjadi pada hari kerja, menunjukkan bahwa pengguna yang terdaftar berperan penting dalam menjaga stabilitas dan peningkatan penggunaan sepeda.

Waktu penggunaan tertinggi di pagi dan sore hari menandakan bahwa sepeda lebih sering digunakan oleh para pekerja. Selain itu, cuaca juga mempengaruhi jumlah penggunaan sepeda, dengan cuaca cerah mendorong lebih banyak penyewaan. Meskipun musim tidak terlalu berpengaruh secara keseluruhan, terdapat penurunan penggunaan sepeda yang signifikan selama musim semi.

Dengan pola ini, bisa disimpulkan bahwa bike sharing lebih banyak dimanfaatkan oleh pekerja pada jam sibuk dan cuaca yang mendukung.
